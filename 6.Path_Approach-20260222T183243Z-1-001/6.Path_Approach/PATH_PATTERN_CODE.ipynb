```python
import networkx as nx

# Create a graph representing the friendship network
G = nx.Graph()
#G.add_edges_from([(1, 2), (1, 3), (2, 4), (3, 5), (5, 6)])
for u,v in zip(source, dest):
    G.add_edge(u, v)

print(G.edges(data=True))

csv ={}
total_weight = 0
# Set the depression statuses of some users
#depression_status = {1: False, 2: True, 3: False, 4: False, 5: True}

# Define the target user and the maximum number of levels to consider
for iu in range(1,251):
  target_user = iu
  max_levels = 3

# Run a shortest path estimator for the whole friendship network
  if target_user not in source:
    continue
  else:
    shortest_paths = nx.shortest_path_length(G, target_user)
    print(shortest_paths)
# Calculate the weight of each neighbor based on its distance from the target user
    weights = {}
    for neighbor, distance in shortest_paths.items():
        if distance <= max_levels:
            weights[neighbor] = max_levels - distance + 1
    print(weights)

# Calculate the possibility of each depressed neighbor affecting the target user
    for key,value in weights.items():
      total_weight = total_weight+value
    possibilities = {}
    for neighbor, distance_weight in weights.items():
        if neighbor in depression_status and depression_status[neighbor]:
            possibilities[neighbor] = distance_weight / total_weight
    print(possibilities)

    for key in possibilities.keys():
      if key == target_user:
        possibilities.pop(target_user)
        break
#print(possibilities)

    sum = 0
    temp = []
    for x in possibilities.values():
      temp.append(round(x,5))
#print(temp)


    for i in temp:
     sum = sum+i
    sum = sum*1

  print("the influence score of user",iu,":",sum)
  csv[iu] = sum




```
